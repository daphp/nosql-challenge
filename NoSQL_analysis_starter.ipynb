## Notebook Set Up

```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
collections = db.list_collection_names()
print("Collections in 'uk_food' database:", collections)
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = db.establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {count}")
# Display the first document in the results using pprint
first_result = db.establishments.find_one(query)
print("\nFirst establishment with a hygiene score of 20:")
pprint(first_result)
```

```python
# Convert the result to a Pandas DataFrame
cursor = db.establishments.find(query)
results = list(cursor)  # Convert cursor to list to load into DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
num_rows = len(df)
print(f"\nNumber of rows in the DataFrame: {num_rows}")
# Display the first 10 rows of the DataFrame
print("\nFirst 10 rows of the DataFrame:")
print(df.head(10))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Define the query to find establishments in London with a RatingValue >= 4
query = {
    "LocalAuthorityName": {"$regex": r"\bLondon\b", "$options": "i"},
    "RatingValue": {"$gte": 4}  # RatingValue greater than or equal to 4
}

# Use count_documents to display the number of documents in the result
count = db.establishments.count_documents(query)
print(f"Number of establishments in London with a RatingValue >= 4: {count}")

# Retrieve the first document in the results
first_document = db.establishments.find_one(query)

# Display the first document in the results using pprint
if first_document:
    print("\nFirst document in the results:")
    pprint(first_document)
else:
    print("No establishments found in London with a RatingValue >= 4.")
```

```python
# Convert the result to a Pandas DataFrame
results = db.establishments.find(query)
df = pd.DataFrame(list(results))
# Display the number of rows in the DataFrame
num_rows = df.shape[0]
print(f"Number of rows in the DataFrame: {num_rows}")
# Display the first 10 rows of the DataFrame
print("\nFirst 10 rows of the DataFrame:")
print(df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Latitude and Longitude of Penang Flavours
latitude = 51.49014200
longitude = 0.08384000

# Defining the degree search range
degree_search = 0.01

# Defining the query to find establishments with RatingValue 5 within the latitude and longitude range
query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}

# Sorting by Hygiene score (ascending order)
sort = [("scores.Hygiene", 1)]

# Limiting the results to the top 5 establishments
limit = 5

# Retrieving the results
results = db.establishments.find(query).sort(sort).limit(limit)

# Printing the results
print("\nTop 5 establishments sorted by Hygiene score:")
for establishment in results:
    pprint({
        "Business Name": establishment['BusinessName'],
        "Address": f"{establishment['AddressLine1']}, {establishment.get('PostCode', 'No PostCode')}",
        "Rating Value": establishment['RatingValue'],
        "Hygiene Score": establishment['scores'].get('Hygiene', 'No Hygiene Score'),
        "Latitude": establishment['geocode']['latitude'],
        "Longitude": establishment['geocode']['longitude']
    })
    print("-" * 40)
```

```python
# Re-running the query to get the top 5 establishments again
results = db.establishments.find(query).sort(sort).limit(limit)

# Converting results to a list of dictionaries
establishment_list = []
for establishment in results:
    establishment_list.append({
        "Business Name": establishment['BusinessName'],
        "Address": f"{establishment['AddressLine1']}, {establishment.get('PostCode', 'No PostCode')}",
        "Rating Value": establishment['RatingValue'],
        "Hygiene Score": establishment['scores'].get('Hygiene', 'No Hygiene Score'),
        "Latitude": establishment['geocode']['latitude'],
        "Longitude": establishment['geocode']['longitude']
    })

# Creating a DataFrame
df_establishments = pd.DataFrame(establishment_list)

# Displaying the DataFrame
print("\nDataFrame of Top 5 establishments sorted by Hygiene score:")
print(df_establishments)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Creating the aggregation pipeline
pipeline = [
    # Step 1: Match establishments with a hygiene score of 0
    {
        "$match": {
            "scores.Hygiene": 0
        }
    },
    # Step 2: Group by Local Authority and count the number of establishments
    {
        "$group": {
            "_id": "$LocalAuthorityName",  # Group by Local Authority
            "count": {"$sum": 1}           # Count the number of establishments
        }
    },
    # Step 3: Sort by count in descending order
    {
        "$sort": {
            "count": -1  # Sort from highest to lowest
        }
    }
]

# Executing the aggregation pipeline
results = db.establishments.aggregate(pipeline)

# Converting results to a list
result_list = list(results)

# Printing the number of documents in the result
print(f"Number of Local Authorities with establishments having a hygiene score of 0: {len(result_list)}")

# Printing the first 10 results
print("\nFirst 10 Local Authorities with hygiene score of 0:")
for authority in result_list[:10]:
    print(f"Local Authority: {authority['_id']}, Count: {authority['count']}")
```

```python
# Convert the result to a Pandas DataFrame
df_results = pd.DataFrame(result_list)
# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df_results)}")
# Display the first 10 rows of the DataFrame
print("\nFirst 10 rows of the DataFrame:")
print(df_results.head(10))
```
